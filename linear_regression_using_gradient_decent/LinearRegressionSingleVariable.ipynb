```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('tvmarketing.csv')
x=data.iloc[:,0]
y=data.iloc[:,1]
type(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def hypothesis(x):
    return m*x+c
epochs=100
lr=0.000001
m=0.1
c=0.1
n = float(len(x))
for i in range(epochs):
    y_hat = hypothesis(x)
    loss = ((y_hat-y)**2).mean()
    temp_m = m-lr*((-2/n)*sum(x*(y-y_hat)))
    temp_c = c-lr*((-2/n)*sum(y-y_hat))
    m=temp_m
    c=temp_c
    if i%10==0:
        print(f'loss={loss:.3f},m={m:.3f},c={m:.3f}')
```

```python
plt.scatter(x,y)
a=np.linspace(0, 300, 100)
b=m*a+c
plt.plot(a,b,'red')
plt.show()
```
